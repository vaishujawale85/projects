```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
```

```python
df=pd.read_csv("C:\\Users\\patil\\Downloads\\creditcard.csv")
print(df)
```

```python
# Check for missing values
print("Missing values:")
print(df.isnull().sum().sum())
```

```python
# Data preprocessing
X = df.drop(columns=['Class'])
y = df['Class']
```

```python
# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Splitting the dataset
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

```python
# Anomaly Detection using Isolation Forest
iso_forest = IsolationForest(n_estimators=100, contamination=0.02, random_state=42)
y_pred_iso = iso_forest.fit_predict(X_test)
y_pred_iso = np.where(y_pred_iso == 1, 0, 1)  # Convert to match fraud labels
```

```python
# Model evaluation
print("Isolation Forest Results:")
print(confusion_matrix(y_test, y_pred_iso))
print(classification_report(y_test, y_pred_iso))
```

```python
# Visualizing the fraud cases
plt.figure(figsize=(10,5))
sns.countplot(x=y, palette='coolwarm')
plt.title("Class Distribution",color="red")
plt.show()
```
